```python
import matplotlib.pyplot as plt
%matplotlib inline 
# evita usar uma linha plt.show()
```

```python
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
x
```

```python
y = x ** 2
y
```

# Usando a matplotlib -- forma simples

```python
plt.plot(x, y, color = 'r')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Titulo')
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r--')
plt.subplot(1, 2, 2)
plt.plot(y, x, 'g*-')
```

# Usando a matplotlib -- forma completa

### Configurando o posicionamento



```python
fig = plt.figure()
axes = fig.add_axes([.1, .1, .8, .8]) #10% de distância da parte esquerda e inferior e 80% de largura e altura

axes.set_xlabel('Eixo X')
axes.set_ylabel('Eixo Y')
axes.set_title('Título')

axes.plot(x, y)
```

```python
fig = plt.figure()
axes1 = fig.add_axes([.1, .1, .8, .8]) #10% de distância da parte esquerda e inferior e 80% de largura e altura

axes1.set_xlabel('Eixo X')
axes1.set_ylabel('Eixo Y')
axes1.set_title('Título')
axes1.plot(x, y)


axes2 = fig.add_axes([.2, .4, .3, .3])
axes2.plot(y, x)

```

# Subplots

```python
fig, ax = plt.subplots()
ax.plot(x, x**3, 'b--')
ax.plot(x, x**4, 'r.-')
ax.set_xlabel('Eixo x')
ax.set_ylabel('Eixo y')
ax.set_title('Título')
```

```python
fig, ax = plt.subplots(nrows = 1, ncols = 2)
ax[0].plot(x, x**2, 'b--')
ax[1].plot(x, x**4, 'r')
```

```python
fig, ax = plt.subplots(nrows = 5, ncols = 5)
plt.tight_layout()
```

# Configurações da Figura

```python
fig, ax = plt.subplots( figsize = ( 12, 3 ), dpi = 100)
ax.plot(x, y, 'r', label = 'y = x')
ax.plot(x, y*4, 'g', label = 'y = x * 4')
ax.set_title('Título')
ax.legend(loc = 2) # loc = posição // o valor padrão é selecionado a melhor posição
```

```python
fig.savefig('imagem.png')
```

# Alterando o Layout

```python
fig , ax = plt.subplots(figsize = (10, 10), dpi = 100)
ax.plot(x, x**2, color = 'red' , label = 'y = x²')
ax.plot(x, x**3, color = 'blue', label = 'y = x³', linewidth = 10, alpha = 0.2, linestyle = ':')
ax.legend(fontsize = 'large')

ax.set_xlim([0, 6])
ax.set_ylim([0, 150])
```
